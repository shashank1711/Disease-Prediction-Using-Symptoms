### Feature Engineering

```python
data.isnull().any().tolist()
```

```python
# Splitting into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =0, test_size= 0.20)
```

```python
print('Training Set: ', len(X_train))
print('Test Set: ', len(X_test))
print('Training Labels: ', len(y_train))
print('Test Labels: ', len(y_test))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
classifier = RandomForestClassifier(n_estimators= 300, max_depth= 8, random_state= 0, criterion= 'entropy')
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print('Classification Report: ')
print(classification_report(y_test, y_pred))
```

```python
print('Accuracy Score: '+ str(accuracy_score(y_test, y_pred)*100)+ ' %')
```

```python
f, ax = plt.subplots(figsize = (18, 15))
cm = confusion_matrix(y_test, y_pred)
cmap = sns.diverging_palette(220, 10, as_cmap = True)
sns.heatmap(cm, cmap = cmap, vmax = 0.3, center = 0,
            square = True, linewidths= 0.5)
```

### Feature Importance

```python
importances = classifier.feature_importances_
indices = np.argsort(importances)[: : -1]
print('Symptoms Ranking: ')
for f in range(X_train.shape[1]):
    print('Symptom %s (%f)' %(list(X_train)[f], importances[indices[f]]))

feat_imp = pd.DataFrame({'Symptoms': list(X_train),
                         'Importance': importances[indices]})
plt.rcParams['figure.figsize'] = (18, 20)
sns.set_style('whitegrid')
ax = sns.barplot(x = 'Importance', y = 'Symptoms', data = feat_imp)
```

```python
# Permutation Importance
import eli5
from eli5.sklearn import PermutationImportance
perm = PermutationImportance(classifier, random_state= 0).fit(X_test, y_test)
eli5.show_weights(perm, feature_names = X_test.columns.tolist())
```
